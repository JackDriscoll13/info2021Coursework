```python
#write a function that takes a list of files and counts all characters in that file:
def countCharactersInFiles(listOfFileNames):
    lengthOfList = len(listOfFileNames) #how many file names in list 
    charCounter = 0 
    for file in listOfFileNames:
        myFile = open(str(file),"r")
        fContents = myFile.read()
        myFile.close
        print("There are are " + str(len(fContents))+ " characters in this file ")
        charCounter += len(fContents)
    
    print("There are "+str(charCounter)+" characters in "+str(lengthOfList)+" files.")
    
```

```python
# a function that calculates the total number of albums an artist has in specificed file
import csv

def findAlbumTotalForAllBands(fileName):
    # we'll need a dictionary to store artist name and num of albums (for each artist)
    albumCount = {} #create empty dictionary
    file = open(fileName, "r")
    csvReader = csv.reader(file)
    header = next(csvReader) #this skips the first line in the file header
    for row in csvReader: #look at each line which is a list of word
        if (row[3] not in albumCount):
            #Add to dictionary #syntax
            #dictionaryname[keyname] = value
            #so 
            albumCount[row[3]]=1
        else:
            albumCount[row[3]]+=1
    file.close()
    return albumCount

#print(findAlbumTotalForAllBands("RollingStoneAlbumList.csv"))
# put in in insertion order 


        
```

```python
#looking at a text file and seeing which words are used the most


def countAllWordInstance(fileName):
    file = open(fileName,"r")
    fileContents = file.read().replace(",","").replace("!","").replace("?","").replace(".","").replace("(","").replace(")","").lower()
    file.close()
    wordDictionary = {}
    wordList = fileContents.split()
    for word in wordList: #looks at each word in file
        if (word not in wordDictionary):
            wordDictionary[word]=1 #add to dictionary
        else:
            wordDictionary[word]+=1
    return wordDictionary
    #for word in wordDictionary:
        #print(word+": "+str(wordDictionary[word]))
beatlesWords = countAllWordInstance("AllBeatlesLyrics.txt")

```
